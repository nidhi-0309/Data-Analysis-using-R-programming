**Recoding variables using ifelse**

Use ifelse() function to create new character variable agecat for ages less than the median 10.5 and ages equal to or older than 10.5, labeled as "< 10.5" and "10.5 and older".
Add agecat to abaloneMod.

```r
abaloneMod <- abaloneMod %>% 
mutate(agecat=ifelse(test=age<10.5,
					yes="<10.5",
					no= ifelse(age>=10.5,"10.5 and older","unknown")))
```

```r
abaloneMod
```

**Add new logical variable adult to abaloneMod**

```r
abaloneMod <- abaloneMod %>% 
mutate(adult= sex != "I")
```

```r
abaloneMod
```

## **Tests for two groups**

```r
# Get n, mean, sd for length by adult groups
abaloneKeep %>% 
  select(length, adult) %>%
  group_by(adult) %>%
  summarise(across(everything(), list(mean = ~ mean(.x),
                                      sd = ~ sd(.x))),
            N = n())
```

```r
# Run equal variance and pooled t-tests of length by adult 
var.test(length~adult,data=abaloneKeep)
t.test(length~adult,data=abaloneKeep ,var.equal = TRUE)
```

```r
# Make boxplots of length by adult
ggplot(abaloneKeep, aes(adult, length)) +
  geom_boxplot()
```

_**Observation:T-tests provide useful insights into group differences. For example, the adult abalones do have longer lengths than the infants, but the difference in group mean lengths is within approximately two standard _deviations.**_
